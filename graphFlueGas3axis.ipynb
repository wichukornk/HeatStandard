```python
import pandas as pd
import plotly.graph_objects as go

path = 'C:/Users/wichu/OneDrive/1. Work/4. BestEN/1. Heat Threshold/1 /InspectionData/FlueGas/29309-0054-1.csv'

df = pd.read_csv(path)

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df['runtime'],
    y=df['flueTemp'],
    mode='lines',
    marker_color='#d62728',
    name="Flue gas temp"
))


fig.add_trace(go.Scatter(
    x=df['runtime'],
    y=df['o2'],
    mode='lines',
    marker_color='#1f77b4',
    name="O2[%]",
    yaxis="y2"
))

fig.add_trace(go.Scatter(
    x=df['runtime'],
    y=df['co'],
    mode='lines',
    marker_color='#ff7f0e',
    name="CO[ppm]",
    yaxis="y3"
))

#-------------------------------------------------------#

fig.update_layout(
    xaxis=dict(
        domain=[0.0, 0.9]
    ),
    yaxis=dict(
        title="Flue gas temp[`C]",
        #range = [180,190],
        showgrid=False,
        titlefont=dict(
            color="#d62728"
        ),
        tickfont=dict(
            color="#d62728"
        )
    ),
    yaxis2=dict(
        title="O2[%]",
        #range = [10,15],
        showgrid=False,
        titlefont=dict(
            color="#1f77b4"
        ),
        tickfont=dict(
            color="#1f77b4"
        ),
        anchor="free",
        overlaying="y",
        side="right",
        position=0.95
    ),
    yaxis3=dict(
        title="CO[ppm]",
        showgrid=False,
        titlefont=dict(
            color="#ff7f0e"
        ),
        tickfont=dict(
            color="#ff7f0e"
        ),
        anchor="x",
        overlaying="y",
        side="right"
    )
)

# Update layout properties
fig.update_layout(
    title_text="Fortune Part Industry",
    width=1600,
)


fig.show()
```
